```python
from cycler import cycler
v = layout.view
fig, ax = plt.subplots(1)
colormap = plt.get_cmap('nipy_spectral')
colors = [colormap(i) for i in np.linspace(0, 1, len(v.elements))]
ax.set_prop_cycle(cycler('color', colors))
for si in v.shapes:
    ax.fill(si.get_x(), si.get_y(), label=si.name)
for ti in v.texts:
    ax.text(ti.get_x(), ti.get_y(), ti.label)
ax.legend(loc='upper right')
```
